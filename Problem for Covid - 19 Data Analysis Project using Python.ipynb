# 4. Filtering the Dataframe 

```python
cov_data_new = cov_data[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
```

```python
cov_data_new.head()
```

# 5. Data Cleaning 

```python
# a. Remove all duplicates observations
cov_data_new = cov_data_new.drop_duplicates()
```

```python
# b. Find missing values in all columns
cov_data_new.isnull().sum()
```

```python
# c. Remove all observations where continent column value is missing
cov_data_new = cov_data_new[cov_data_new['continent'].notna()]
```

```python
# d. Fill all missing values with 0
cov_data_new = cov_data_new.fillna(0)
```

# 6. Date time format 

```python
# a. Convert date column in datetime format using pandas.to_datetime
cov_data_new.head()
```

```python
cov_data_new['date'] = pd.to_datetime(cov_data_new['date']) 
```

```python
cov_data_new.head()
```

```python
# b. Create new column month after extracting month data from date column
cov_data_new['Month'] = cov_data_new['date'].dt.month
```

```python
cov_data_new.head()
```

# 7. Data Aggregation

```python
# Find max value in all columns using groupby function on 'continent' column
cov_data_new.groupby(['continent']).max().reset_index()
```

```python
#  b. Store the result in a new dataframe named 'df_groupby'.
df_groupby = cov_data_new.groupby(['continent']).max().reset_index()
```

# 8. Feature Engineering

```python
# a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths']/df_groupby['total_cases'] 
```

```python
df_groupby.head()
```

# 9. Data Visualization 

```python
# a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
import seaborn as sns

sns.distplot(df_groupby['gdp_per_capita'].values)
```

```python
#  b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
import matplotlib.pyplot as plt
plt.scatter(df_groupby['total_cases'],df_groupby['gdp_per_capita'])
```

```python
# c. Plot Pairplot on df_groupby dataset
sns.pairplot(df_groupby)
```

```python
#  d. Plot a bar plot of 'continent' column with 'total_cases'
plt.bar(df_groupby['continent'],df_groupby['total_cases'])
plt.show()
```
